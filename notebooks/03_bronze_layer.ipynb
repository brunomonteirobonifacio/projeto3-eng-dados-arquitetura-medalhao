#1. Carregar os dados da Landing Zone

```python
# Lê as tabelas da camada Landing
df_estado          = spark.table("workspace.landing.estado")
df_cidade          = spark.table("workspace.landing.cidade")
df_usuario         = spark.table("workspace.landing.usuario")
df_ouvidoria       = spark.table("workspace.landing.ouvidoria")
df_tipo_ouvidoria  = spark.table("workspace.landing.tipo_ouvidoria")
df_servico_afetado = spark.table("workspace.landing.servico_afetado")
df_anexo           = spark.table("workspace.landing.anexo")
```

# 2. Adicionar colunas de Metadados (Data/Hora e Origem)

```python
from pyspark.sql.functions import current_timestamp, lit

# Adiciona a data/hora e a origem dos dados
df_estado          = df_estado.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.estado"))
df_cidade          = df_cidade.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.cidade"))
df_usuario         = df_usuario.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.usuario"))
df_ouvidoria       = df_ouvidoria.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.ouvidoria"))
df_tipo_ouvidoria  = df_tipo_ouvidoria.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.tipo_ouvidoria"))
df_servico_afetado = df_servico_afetado.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.servico_afetado"))
df_anexo           = df_anexo.withColumn("data_hora_bronze", current_timestamp()).withColumn("fonte_dados", lit("landing.anexo"))
```

#3. Salvar no schema Bronze

```python
# Salva os dataframes na camada Bronze
df_estado.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.estado")
df_cidade.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.cidade")
df_usuario.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.usuario")
df_ouvidoria.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.ouvidoria")
df_tipo_ouvidoria.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.tipo_ouvidoria")
df_servico_afetado.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.servico_afetado")
df_anexo.write.format('delta').mode("overwrite").saveAsTable("workspace.bronze.anexo")

print("Todas as 7 tabelas foram processadas e salvas na camada Bronze com sucesso!")
```
